```python
env = gym.make("Taxi-v3")
```

```python
action_size = env.action_space.n
state_size = env.observation_space.n

print ('State Space : ',state_size)
print ('Action space : ',action_size)
```

```python
# Build simple RL
```

```python
total_episodes = 8000        
max_steps = 90               
gamma = 0.99                 

# Exploration parameters
epsilon = 1.0                
max_epsilon = 1.0            
min_epsilon = 0.01            
decay_rate = 0.005           
learning_rate = 0.85

qtable = np.zeros((state_size, action_size)) 
print(qtable)
```

```python
# List of rewards
rewards = []

# 2 For life or until learning is stopped
for episode in range(total_episodes):
    # Reset the environment
    state = env.reset()
    step = 0
    done = False
    total_rewards = 0
    
    for step in range(max_steps):
        explore_exploit = random.uniform(0, 1)
        
        if explore_exploit > epsilon:
            action = np.argmax(qtable[state,:])

        else:
            action = env.action_space.sample()
        
        new_state, reward, done, info = env.step(action)

        # Update Q(s,a):= Q(s,a) + lr [R(s,a) + gamma * max Q(s',a') - Q(s,a)]
        # qtable[new_state,:] : all the actions we can take from new state
        qtable[state, action] = qtable[state, action] + learning_rate * (reward + gamma * np.max(qtable[new_state, :]) - qtable[state, action])
        
        total_rewards += reward
                
        state = new_state
        
        if done == True: 
            break
        
    # We reduce epsilon because we don't want the agent just explore over and over again but never exploit
    epsilon = min_epsilon + (max_epsilon - min_epsilon)*np.exp(-decay_rate*episode) 
    rewards.append(total_rewards)

print ("Score over time: " +  str(sum(rewards)/total_episodes))
#print(qtable)
```

```python
"""Display and evaluate agent's performance after Q-learning."""
env.reset()
display_episodes=10

total_epochs, total_penalties = 0, 0

for _ in range(display_episodes):
    state = env.reset()
    epochs, penalties, reward = 0, 0, 0
    
    done = False
    
    while not done:
        action = np.argmax(qtable[state])
        state, reward, done, info = env.step(action)

        if reward == -10:
            penalties += 1

        epochs += 1
        env.render()
        print(f"Timestep: {epochs}")
        print(f"State: {state}")
        print(f"Action: {action}")
        print(f"Reward: {reward}")
        sleep(0.15) # Sleep so the user can see the 

    total_penalties += penalties
    total_epochs += epochs

env.close()

print(f"Results after {display_episodes} episodes:")
print(f"Average timesteps per episode: {total_epochs / display_episodes}")
print(f"Average penalties per episode: {total_penalties / display_episodes}")
```
